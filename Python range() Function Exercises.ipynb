```python
# Create a range from 0 to 50, excluding 50.
for x in range(50):
    print(x, end=" - ")
```

```python
# Create a range from 0 to 10 with steps of 2.
for x in range(0,10,2):
    print(x, end=" - ")
```

```python
# Create a range from 100 to 160 with steps of 10. Then print it as a list.
lst = []
for x in range(100,160,10):
    lst.append(x)
print(lst)

# OR

rng = range(100,160,10)
print(list(rng))
```

```python
# Can you create a list from 1300 to 700 with descending steps of 100?
# Is your stop value included in the list?

rng = range(1300,700,-100)
print(list(rng))
```

```python
# Can you you create a list from 1300 to 700 with descending steps of 100, this time including 700?
rng = range(1300,700-1,-100)
print(list(rng))
```

```python
# The aim of this exercise is to count down as follows:
import time
for x in range(9,0,-1):
    time.sleep(1) # to take delay of 1 sec
    print(f"Coundown ... {x}", end="\r")
print("COUNTDOWN DONE !!", end="\r")
```

```python
# Prime Numbers:
for x in range(50):
    if x > 1:
        if x % 2 != 0:
            for y in range(2,x):
                if x / y != 1:
                    print(x)
```
